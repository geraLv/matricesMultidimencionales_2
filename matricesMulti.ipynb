#Matrices multidimencionales

```python
users = []  

def mostrar_menu():  
    print("Menu: seleccione la opción que desea realizar.")  
    print("1 - Crear un nuevo usuario")  
    print("2 - Mostrar todos los datos")  
    print("3 - Filtrar por DNI")  
    print("4 - Salir del programa")  
    return input("Seleccione una opción: ")  

def registrar():  
    nombre = input("Ingrese el nombre del usuario: ")  
    apellido = input("Ingrese el apellido del usuario: ")  
    dni = input("Ingrese el DNI del usuario: ")  

    # Ingresar números de teléfono  
    telefono = input("Ingrese el/los números de teléfono separados por una coma (Ejemplo: 3704, 3707): ")  
    nums_asociados = telefono.split(", ")  

    # Ingresar nombres de hijos  
    nombre_hijo = input("Ingrese el/los nombres de sus hijos separados por una coma (Ejemplo: Pedro, Maria): ")  
    hijos = nombre_hijo.split(", ")  

    # Agregar el usuario a la lista  
    users.append({  
        "nombre": nombre,  
        "apellido": apellido,  
        "dni": dni,  
        "nums_asociados": nums_asociados,  
        "hijos": hijos,  
    })  
    
    print("Usuario agregado correctamente.")  

def mostrar_usuarios():  
    if users:  
        for user in users:  
            print(f"Nombre: {user['nombre']}, Apellido: {user['apellido']}, DNI: {user['dni']}, "  
                  f"Números de teléfono: {', '.join(user['nums_asociados'])}, Hijos: {', '.join(user['hijos'])}")  
    else:  
        print("No hay usuarios registrados.")  

def filtrar_por_dni():  
    dni_buscar = input("Ingrese el DNI que desea filtrar: ")  
    user_filtered = [user for user in users if user['dni'] == dni_buscar]  
    
    if user_filtered:  
        for user in user_filtered:  
            print(f"Nombre: {user['nombre']}, Apellido: {user['apellido']}, "  
                  f"Números de teléfono: {', '.join(user['nums_asociados'])}, Hijos: {', '.join(user['hijos'])}")  
    else:  
        print("No se encontraron usuarios con ese DNI.")  

def main():  
    while True:  
        opcion = mostrar_menu()  
        
        if opcion == "1":  
            registrar()  
        elif opcion == "2":  
            mostrar_usuarios()  
        elif opcion == "3":  
            filtrar_por_dni()  
        elif opcion == "4":  
            break  
        else:  
            print("Opción no válida. Por favor, seleccione una opción del menú.")  

if __name__ == "__main__":  
    main()  
```
